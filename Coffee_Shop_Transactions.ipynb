```python
#Import necessary modules into the environment

import os
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
```

```python
#Import file into the environment

uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])

print(df.head())
```

```python
#Get the data frame stats and go through each column

df.describe()
```

```python
#First we'll make a dictionary of food and drinks and we'll see which is more popular
food = {
    "chicken nuggets": df["Crispy Chicken Nuggets"],
    "fries": df["Fries"],
   "egg salad toast": df["Egg Salad Toast"],
    "croissant": df["Croissant"],
    "cookies": df["Cookies"],
    "butter toast": df["Butter Toast"],
    "choc toast": df["Chocolate Toast"],
    "cheese toast": df["Cheese Toast"],
    "matcha bun": df["Matcha Bun"]
}

drink = {
    "latte": df["Latte"],
    "hazelnut latte": df["Hazelnut Latte"],
    "vanilla latte": df["Vanilla Latte"],
    "creamy aren latte": df["Creamy Aren Latte"],
    "caramel latte": df["Caramel Latte"],
    "matcha latte": df["Matcha Latte"],
    "cappuccino": df["Cappuccino"],
    "caramel macchiato": df["Caramel Macchiato"],
    "americano": df["Americano"]
}
```

```python
#Determine total sales for food
food_sales = {} # Create a dictionary to store the total sales for each food item
for item, values in food.items():
  total_sales = values.sum()
  food_sales[item] = total_sales # Store the total sales for the current item
  print(f"Total sales for {item}:{total_sales}")

#Bar Graph of each item sales

plt.figure(figsize = (8, 6))
plt.bar(food_sales.keys(), food_sales.values())
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.xlabel("Food Items")
plt.ylabel("Number of Transactions")
plt.title("Total Food Sales")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
total_drink_sales = {}

for drink_name, drink_data in drink.items():
  drink_sales = drink_data.sum()
  total_drink_sales[drink_name] = drink_sales  # Store the total sales using drink_name as the key
  print(f"Total sales for {drink_name}:{drink_sales}")

plt.figure(figsize=(8, 6))
plt.bar(total_drink_sales.keys(), total_drink_sales.values())
plt.xlabel("Drink Items")
plt.ylabel("Number of Transactions")
plt.title("Total Drink Sales")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
#Food Versus Drink Popularity
overall_food_sales = sum(pd.to_numeric(list((food_sales.values()))))
overall_drink_sales = sum(pd.to_numeric(list(total_drink_sales.values())))

overall_sales = {"overall_food_sales" : sum(pd.to_numeric(list((food_sales.values())))),
                "overall_drink_sales" : sum(pd.to_numeric(list(total_drink_sales.values())))}

plt.figure(figsize = (8,6))
plt.bar(overall_sales.keys(), overall_sales.values())
plt.title("Food Versus Drink Sales")
plt.xlabel("Food and Drinks")
plt.ylabel("Number of Transactions")
plt.xticks(rotation = 45, ha = "right")
plt.show()
```

From the previous bar charts, we can determine the following:


1. Even though food sells more than drinks, that difference is very small, with total food sales being 1503 and total drink sales being 1460.

2.   The Most Popular items are
* Food: croissant
* Drink: hazelnut latte & caramel macchiatto

3. And the Least Popular items are:
* Food: cookies
* Drink: latte
